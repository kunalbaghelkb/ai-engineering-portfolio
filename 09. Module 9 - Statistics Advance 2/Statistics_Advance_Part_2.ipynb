###1. What is hypothesis testing in statistics?
Hypothesis testing is a method to make decisions about a population using sample data.

###2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
The null hypothesis states no effect; the alternative states an effect exists.

###3. What is the significance level in hypothesis testing, and why is it important?
The significance level (α) is the threshold for rejecting the null hypothesis.

###4. What does a P-value represent in hypothesis testing?
It represents the probability of observing results as extreme as the sample.

###5. How do you interpret the P-value in hypothesis testing?
If P ≤ α, reject the null hypothesis; otherwise, fail to reject it.

###6. What are Type 1 and Type 2 errors in hypothesis testing?
Type 1 rejects a true null; Type 2 fails to reject a false null.

###7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
One-tailed tests one direction; two-tailed tests both directions.

###8. What is the Z-test, and when is it used in hypothesis testing?
Z-test is used when population variance is known and sample size is large.

###9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
Z = (X − μ) / σ; it shows distance from mean in standard deviations.

###10. What is the T-distribution, and when should it be used instead of the normal distribution?
T-distribution is used for small samples with unknown variance.

###11. What is the difference between a Z-test and a T-test?
Z-test uses normal distribution; T-test uses t-distribution.

###12. What is the T-test, and how is it used in hypothesis testing?
T-test compares means when sample size is small.

###13. What is the relationship between Z-test and T-test in hypothesis testing?
Both test means; T-test converges to Z-test as sample size increases.

###14. What is a confidence interval, and how is it used to interpret statistical results?
It provides a range likely to contain the population parameter.

###15. What is the margin of error, and how does it affect the confidence interval?
It determines the width of the confidence interval.

###16. How is Bayes' Theorem used in statistics, and what is its significance?
It updates probabilities based on new evidence.

###17. What is the Chi-square distribution, and when is it used?
It is used for categorical data and goodness-of-fit tests.

###18. What is the Chi-square goodness of fit test, and how is it applied?
It checks how well observed data matches expected frequencies.

###19. What is the F-distribution, and when is it used in hypothesis testing?
It is used to compare variances and in ANOVA tests.

###20. What is an ANOVA test, and what are its assumptions?
ANOVA compares multiple means assuming normality, independence, and equal variances.

###21. What are the different types of ANOVA tests?
One-way, two-way, and repeated measures ANOVA.

###22. What is the F-test, and how does it relate to hypothesis testing?
F-test compares variances to test mean differences.

```python
#1 - Perform a Z-test comparing a sample mean to a known population mean.

import numpy as np
from scipy.stats import norm

sample = np.array([52, 55, 50, 53, 54])
mu = 50
sigma = 5
z = (sample.mean() - mu) / (sigma / np.sqrt(len(sample)))
p = 2 * (1 - norm.cdf(abs(z)))
print(z, p)
```

```python
#2 - Simulate random data and perform hypothesis testing with P-value.

import numpy as np
from scipy.stats import norm

data = np.random.normal(100, 10, 50)
z = (data.mean() - 100) / (10 / np.sqrt(len(data)))
print(1 - norm.cdf(abs(z)))
```

```python
#4 - Two-tailed Z-test and visualize decision region.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

z_crit = norm.ppf(0.975)
x = np.linspace(-4,4,400)
plt.plot(x, norm.pdf(x))
plt.axvline(-z_crit)
plt.axvline(z_crit)
plt.show()

print(z_crit)
```

```python
#5 - Visualize Type 1 and Type 2 errors.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4,4,400)
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.5, 1))
plt.show()

print("Type 1 and Type 2 regions visualized")
```

```python
#11 - Bayesian inference using Bayes' Theorem.

prior = 0.01
likelihood = 0.95
false_positive = 0.05

posterior = (likelihood * prior) / ((likelihood * prior) + (false_positive * (1 - prior)))
print(posterior)
```

```python
#13 - Calculate expected frequencies for Chi-square test.

import numpy as np

observed = np.array([[10,20],[20,40]])
row_sum = observed.sum(axis=1)
col_sum = observed.sum(axis=0)
total = observed.sum()
expected = np.outer(row_sum, col_sum) / total

print(expected)
```

```python
#20 - Two-way ANOVA with interaction.

import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd

df = pd.DataFrame({
    "score": np.random.rand(30),
    "A": ["X","Y","Z"]*10,
    "B": ["M","N"]*15
})

model = ols("score ~ A * B", data=df).fit()
print(sm.stats.anova_lm(model))
```
