###1. What is statistics, and why is it important?
Statistics is the science of collecting and analyzing data to make informed decisions.

###2. What are the two main types of statistics?
Descriptive statistics and inferential statistics.

###3. What are descriptive statistics?
They summarize and describe data using measures like mean, median, and charts.

###4. What is inferential statistics?
It draws conclusions about a population using sample data.

###5. What is sampling in statistics?
Sampling is selecting a subset of a population for analysis.

###6. What are the different types of sampling methods?
Random, stratified, systematic, cluster, and convenience sampling.

###7. What is the difference between random and non-random sampling?
Random sampling gives equal chance to all; non-random does not.

###8. Define and give examples of qualitative and quantitative data.
Qualitative data is categorical (gender); quantitative data is numerical (age).

###9. What are the different types of data in statistics?
Qualitative (categorical) and quantitative (numerical).

###10. Explain nominal, ordinal, interval, and ratio levels of measurement.
They represent data categories from no order to meaningful zero.

###11. What is the measure of central tendency?
It represents the central or typical value of a dataset.

###12. Define mean, median, and mode.
Mean is average, median is middle value, mode is most frequent value.

###13. What is the significance of the measure of central tendency?
It provides a single value to summarize data behavior.

###14. What is variance, and how is it calculated?
Variance measures data spread as the average squared deviation from the mean.

###15. What is standard deviation, and why is it important?
It is the square root of variance and shows data dispersion clearly.

###16. Define and explain the term range in statistics.
Range is the difference between maximum and minimum values.

###17. What is the difference between variance and standard deviation?
Standard deviation is the square root of variance and easier to interpret.

###18. What is skewness in a dataset?
Skewness measures the asymmetry of a data distribution.

###19. What does it mean if a dataset is positively or negatively skewed?
Positive skew has a right tail; negative skew has a left tail.

###20. Define and explain kurtosis.
Kurtosis measures the peakedness or tail heaviness of a distribution.

###21. What is the purpose of covariance?
Covariance shows how two variables move together.

###22. What does correlation measure in statistics?
Correlation measures the strength and direction of a relationship.

###23. What is the difference between covariance and correlation?
Correlation is standardized; covariance is not scale-independent.

###24. What are some real-world applications of statistics?
Business forecasting, healthcare analysis, finance, and scientific research.

```python
#1 - How do you calculate the mean, median, and mode of a dataset?

import numpy as np
from scipy import stats

data = [10, 20, 20, 30, 40]
print(np.mean(data), np.median(data), stats.mode(data, keepdims=True).mode[0])
```

```python
#2 - Write a Python program to compute the variance and standard deviation of a dataset.

import numpy as np

data = [10, 20, 30, 40]
print(np.var(data), np.std(data))
```

```python
#3 - Create a dataset and classify it into nominal, ordinal, interval, and ratio types.

nominal = ["Male", "Female"]
ordinal = ["Low", "Medium", "High"]
interval = [10, 20, 30]
ratio = [1, 2, 3]

print(nominal, ordinal, interval, ratio)
```

```python
#4 - Implement sampling techniques like random sampling and stratified sampling.

import pandas as pd

df = pd.DataFrame({"group": ["A","A","B","B"], "value":[1,2,3,4]})
print(df.sample(2))
print(df.groupby("group").sample(1))
```

```python
#5 - Write a Python function to calculate the range of a dataset.

def data_range(data):
    return max(data) - min(data)

print(data_range([5, 10, 20, 30]))
```

```python
#6 - Create a dataset and plot its histogram to visualize skewness.

import matplotlib.pyplot as plt

data = [1,2,2,3,4,10]
plt.hist(data)
plt.show()
```

```python
#7 - Calculate skewness and kurtosis of a dataset using Python libraries.

from scipy.stats import skew, kurtosis

data = [1,2,2,3,4,10]
print(skew(data), kurtosis(data))
```

```python
#8 - Generate a dataset and demonstrate positive and negative skewness.

pos_skew = [1,2,2,3,10]
neg_skew = [10,8,6,6,5]

print(skew(pos_skew), skew(neg_skew))
```

```python
#9 - Write a Python script to calculate covariance between two datasets.

import numpy as np

x = [1,2,3]
y = [4,5,6]
print(np.cov(x, y))
```

```python
#10 - Write a Python script to calculate the correlation coefficient between two datasets.

import numpy as np

x = [1,2,3]
y = [4,5,6]
print(np.corrcoef(x, y))
```

```python
#11 - Create a scatter plot to visualize the relationship between two variables.

import matplotlib.pyplot as plt

plt.scatter(x, y)
plt.show()
```

```python
#12 - Implement and compare simple random sampling and systematic sampling.

import pandas as pd

df = pd.DataFrame({"num": range(1,11)})
print(df.sample(3))
print(df.iloc[::3])
```

```python
#13 - Calculate the mean, median, and mode of grouped data.

import pandas as pd

data = pd.Series([10,20,20,30])
print(data.mean(), data.median(), data.mode()[0])
```

```python
#14 - Simulate data using Python and calculate its central tendency and dispersion.

import numpy as np

data = np.random.normal(50, 10, 100)
print(np.mean(data), np.std(data))
```

```python
#15 - Use NumPy or pandas to summarize a dataset's descriptive statistics.

import pandas as pd

df = pd.DataFrame({"values":[10,20,30,40]})
print(df.describe())
```

```python
#17 - Calculate the interquartile range (IQR) of a dataset.

import numpy as np

data = [10,20,30,40,50]
q1, q3 = np.percentile(data, [25,75])
print(q3 - q1)
```

```python
#18 - Implement Z-score normalization and explain its significance.

import numpy as np

data = [10,20,30]
z_scores = (data - np.mean(data)) / np.std(data)
print(z_scores)
```

```python
#19 - Compare two datasets using their standard deviations.

import numpy as np

a = [10,20,30]
b = [10,50,90]
print(np.std(a), np.std(b))
```
