###1. What is NumPy, and why is it widely used in Python?
NumPy is a numerical computing library that offers fast array operations and mathematical functions.

###2. How does broadcasting work in NumPy?
Broadcasting allows operations on arrays of different shapes without explicit loops.

###3. What is a Pandas DataFrame?
A DataFrame is a 2D labeled tabular data structure with rows and columns.


###4. Explain the use of the groupby() method in Pandas.
groupby() splits data into groups and applies aggregation or transformations.

###5. Why is Seaborn preferred for statistical visualizations?
Seaborn provides high-level statistical plots with clean and informative defaults.

###6. What are the differences between NumPy arrays and Python lists?
NumPy arrays are faster and fixed-type; lists are flexible but slower.

###7. What is a heatmap, and when should it be used?
A heatmap uses color to show data intensity, useful for correlations and patterns.

###8. What does the term "vectorized operation" mean in NumPy?
It means applying operations on entire arrays at once without loops.

###9. How does Matplotlib differ from Plotly?
Matplotlib is static; Plotly is interactive and web-based.

###10. What is the significance of hierarchical indexing in Pandas?
It enables multi-level indexing for complex data slicing and grouping.

###11. What is the role of Seaborn's pairplot() function?
pairplot() shows pairwise relationships and distributions between variables.

###12. What is the purpose of the describe() function in Pandas?
describe() provides summary statistics like mean, std, and percentiles.

###13. Why is handling missing data important in Pandas?
Missing values can distort analysis and cause model errors.

###14. What are the benefits of using Plotly for data visualization?
Plotly offers interactive, zoomable, and dashboard-ready charts.

###15. How does NumPy handle multidimensional arrays?
It uses n-dimensional ndarray objects with efficient indexing.

###16. What is the role of Bokeh in data visualization?
Bokeh creates interactive visualizations for web applications.

###17. Explain the difference between apply() and map() in Pandas.
map() works element-wise on Series; apply() works across rows or columns.

###18. What are some advanced features of NumPy?
Broadcasting, linear algebra, FFT, random modules, and memory views.

###19. How does Pandas simplify time series analysis?
It supports datetime indexing, resampling, and rolling operations.

###20. What is the role of a pivot table in Pandas?
It summarizes and aggregates data across categorical dimensions.

###21. Why is NumPy's array slicing faster than Python's list slicing?
NumPy slicing returns views and is optimized in C.

###22. What are some common use cases for Seaborn?
EDA, correlation analysis, distribution plots, and regression visualization.

```python
#1 - How do you create a 2D Numpy array and calculate the sum of each row?

import numpy as np
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr.sum(axis=1))
```

```python
#2 - Write a Pandas script to find the mean of a specific column in a DataFrame.

import pandas as pd
df = pd.DataFrame({"marks": [80, 90, 100]})
print(df["marks"].mean())
```

```python
#3 - Create a scatter plot using Matplotlib.

import matplotlib.pyplot as plt
x = [1, 2, 3]
y = [4, 5, 6]
plt.scatter(x, y)
plt.show()
```

```python
#4 - How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import seaborn as sns
import pandas as pd
df = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})
print(df.corr())
sns.heatmap(df.corr(), annot=True)
```

```python
#5 - Generate a bar plot using Plotly.

import plotly.express as px
df = {"x": ["A", "B", "C"], "y": [10, 20, 30]}
fig = px.bar(df, x="x", y="y")
fig.show()
```

```python
#6 - Create a DataFrame and add a new column based on an existing column.

import pandas as pd
df = pd.DataFrame({"salary": [1000, 2000]})
df["bonus"] = df["salary"] * 0.1
print(df)
```

```python
#7 - Write a program to perform element-wise multiplication of two NumPy arrays.

import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a * b)
```

```python
#8 - Create a line plot with multiple lines using Matplotlib.

import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [2, 4, 6])
plt.plot([1, 2, 3], [1, 3, 5])
plt.show()
```

```python
#9 - Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

import pandas as pd
df = pd.DataFrame({"age": [18, 25, 30]})
print(df[df["age"] > 20])
```

```python
#10 - Create a histogram using Seaborn to visualize a distribution.

import seaborn as sns
data = [10, 20, 20, 30, 30, 30]
sns.histplot(data)
```

```python
#11 - Perform matrix multiplication using NumPy.

import numpy as np
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(a @ b)
```
