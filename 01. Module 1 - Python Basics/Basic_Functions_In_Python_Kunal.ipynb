### 1. What is Python, and why is it popular?
Python is a high-level, interpreted programming language known for its clear syntax and versatility. It's popular due to its ease of learning, extensive libraries (for web, data science, AI, etc.), strong community support, and cross-platform compatibility.

### 2. What is an interpreter in Python?
In Python, an interpreter reads and executes your code line by line, translating it directly into machine instructions.

### 3. What are pre-defined keywords in Python?
Pre-defined keywords are if, else, True, False, global, while, and, def, return.

### 4. Can keywords be used as variable names?
No, Pre-defined keywords in Python are reserved words that have special meanings and cannot be used as variable names, function names, or any other identifiers.

### 5. What is mutability in Python?
Mutability refers to an object's ability to be changed after it's created. Mutable objects (like lists, dictionaries, sets) can have their content altered without changing their identity.

### 6. Why are lists mutable, but tuples are immutable?
Lists are mutable because their internal structure allows elements to be added, removed, or changed after the list has been created. Tuples, on the other hand, are immutable because once they are created, their size and the elements they contain cannot be altered.

### 7. What is the difference between "==" and "is" operators in Python?
The "==" operator checks if two objects have the same value. The "is" operator checks if two variables point to the exact same object in memory.

### 8. What are logical operators in Python?
Logical operators are used to combine conditional statements.

### 9. What is type casting in Python?
Type casting is the process of converting a value from one data type to another.

### 10. What is the difference between implicit and explicit type casting?
Implicit Type Casting happens automatically without explicitly telling Python to do while Explicit Type Casting happens when explicitly tell Python to convert a value from one data type to another using built-in functions.

### 11. What is the purpose of conditional statements in Python?
Conditional statements are used to execute different blocks of code based on whether a specified condition evaluates to True or False.

### 12. How does the elif statement work?
The elif statement is used in conjunction with the if statement to handle multiple conditions.

### 13. What is the difference between for and while loops?
For loop: Runs a set number of times, best when you know the exact number of iterations.
While loop: Runs until a condition becomes false; best when the number of iterations is unknown.

### 14. Describe a scenario where a while loop is more suitable than a for loop.
When waiting for user input (e.g., keep asking for a password until it's correct).

```python
#13. Write a Python program to find the sum of all even numbers between 1 and 50.
total = 0
for i in range(1, 51):
    if i % 2 == 0:
        total += i
print("Sum =", total)
```

```python
#14. Write a program to reverse a string using a while loop.
s = input("Enter a string: ")
rev = ""
i = len(s) - 1
while i >= 0:
    rev += s[i]
    i -= 1
print("Reversed:", rev)
```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
n = int(input("Enter a number: "))
fact = 1
while n > 0:
    fact *= n
    n -= 1
print("Factorial =", fact)
```
